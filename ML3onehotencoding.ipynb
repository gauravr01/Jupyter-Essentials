```python
import pandas as pd

# Define data for the dataframe
data = {'town': ['Town A', 'Town B', 'Town C']*10,
        'area': [100*i for i in range(1,11)] + [200*i for i in range(1,11)] + [300*i for i in range(1,11)],
        'price': [100000*i for i in range(1,11)] + [200000*i for i in range(1,11)] + [300000*i for i in range(1,11)]}

# Create pandas dataframe
df = pd.DataFrame(data)

# Display the dataframe

df.head(10)
```

```python
df['town'].replace({'Town A': 'Mumbai', 'Town B': 'Ahmedabad', 'Town C': 'Goa'}, inplace=True)
df
```

```python
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis = "columns")
merged
```

```python
final = merged.drop(['town','Mumbai'],axis="columns")
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price',axis="columns")
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[1400,1,0]])
```

```python
model.score(X,y) #Accuracy
```

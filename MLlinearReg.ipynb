```python
#MultiVariete Regression with multiple var
import pandas as pd
import numpy as np
# Create the DataFrame
data = pd.DataFrame({
    'area': [1000, 1500, 1200, 1800, 900],
    'bedroom': [2 ,np.nan, 2, 4, 1],
    'age': [10, 5, 8, 12, 15],
    'price': [200000, 300000, 250000, 350000, 180000]
})

# Save the DataFrame to a CSV file
data.to_csv('data.csv', index=False)
data
```

```python
import math
midbed=math.floor(data.bedroom.median())
midbed
```

```python
data.bedroom = data.bedroom.fillna(midbed)
data
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[['area','bedroom','age']],data.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[1550,3,20]])
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
day = [1,2,3,4,5]
min = [23,25,26,27,22]
mix=[55,66,77,88,55]
avg = [12,13,14,17,13]
```

```python
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Weather")
plt.plot(day,min,label='min')
plt.plot(day,mix,label='mix')
plt.plot(day,avg,label='avg')
plt.legend()
plt.grid()
```

```python
a = ['Google','Amazon','Apple','Meta']
b=[333,222,266,239]
profit=[122,106,155,199]
ypos = np.arange(len(a))
ypos

plt.xticks(ypos,a) 
plt.bar(ypos,b)
```

```python
plt.ylabel("Revenue")
plt.title("US Stocks")
plt.bar(ypos-0.2,b,width = 0.4,label="Revenue")
plt.bar(ypos+0.2,profit,width = 0.4,label="Profit")

plt.legend()
```
